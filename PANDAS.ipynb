<h1 style="font-size: 31px;">1. What is Pandas</h1>

Pandas is an open-source Python library for data manipulation and analysis, offering powerful data structures like Series (one-dimensional) and DataFrame (two-dimensional) for handling structured data efficiently.

It simplifies tasks such as data cleaning, transformation, and analysis by providing functions for filtering, merging, and handling missing values. Built on top of NumPy, Pandas ensures high performance and flexibility, making it an essential tool for data science and machine learning applications.


<h1 style="font-size: 31px;">Introduction to NumPy</h1>


## What is NumPy ?

NumPy (Numerical Python) is a fundamental library for numerical computing in Python. It enables efficient handling of large datasets using arrays and provides optimized functions for mathematical operations.

## Why Use NumPy ?

* Faster computations than Python lists
* Supports multi-dimensional arrays for structured data
* Provides built-in mathematical functions for easy calculations
* Optimized memory usage for handling large datasets
* Widely used in data science and machine learning

## Importing NumPy

```python
import numpy as np
```

## Creating a NumPy Array

A NumPy array is a collection of values stored in an efficient array structure:

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

## Basic Operations with NumPy

NumPy provides various operations, such as:

- mean(), median(), std() for statistical calculations

- reshape() to modify array dimensions

- linspace() to create evenly spaced values

```python
# Creating an array
arr = np.array([20, 40, 60, 80, 10])
```

```python
# Performing basic arithmetic
print("Array + 8:", arr + 8)  # Adds 8 to each element  
print("Array * 5:", arr * 5)  # Multiplies each element by 5  
print("Mean of array:", np.mean(arr))  # Calculates mean  
print("Standard Deviation:", np.std(arr))  # Calculates standard deviation  
```

<h1 style="font-size: 31px;">14. Grouping and Aggregation</h1>

## Grouping by 'genres' and calculating average 'revenue'

```python
grouped_df = df.groupby('genres')['revenue'].mean()
print("Average revenue by genres:")
print(grouped_df)
```

## Grouping by 'genres' and calculating median 'revenue'

```python
grouped_df = df.groupby('genres')['revenue'].median()
print("Median revenue by genres:")
print(grouped_df)
```

## Grouping by 'genres' and calculating variance of 'revenue'

```python
grouped_df = df.groupby('genres')['revenue'].var()
print("Variance of revenue by genres:")
print(grouped_df)
```

## Grouping by 'genres' and calculating standard deviation of 'revenue'

```python
grouped_df = df.groupby('genres')['revenue'].std()
print("Standard Deviation of revenue by genres:")
print(grouped_df)
```

<h1 style="font-size: 31px;">15. Renaming a Column</h1>

## Renaming the Column 'title' to 'movie_title'

```python
df = df.rename(columns={"title": "movie_title"})
# Renaming the column 'title' to 'movie_title' in the DataFrame
# The columns parameter is used to specify the old and new column names
# The change is temporary unless reassigned to df
```

```python
df         
# Displaying the DataFrame
```

<h1 style="font-size: 31px;">16. Resetting and Setting Index</h1>

## Setting a Multi-Level Index

```python
df = df.set_index(["movie_title", "vote_average"])
# Setting 'movie_title' and 'vote_average' as the new multi-level index of the DataFrame
# This helps in organizing data based on movie titles and their average ratings
# The change is temporary unless reassigned to df

```

```python
df
# Displaying the DataFrame
```

## Resetting the Index

```python
df = df.reset_index()
# Resetting the index of the DataFrame, converting multi-level indices back into columns
# This restores 'movie_title' and 'vote_average' as regular columns if they were previously set as an index
# The change is temporary unless reassigned to df
```

```python
df 
# Displaying the DataFrame
```
